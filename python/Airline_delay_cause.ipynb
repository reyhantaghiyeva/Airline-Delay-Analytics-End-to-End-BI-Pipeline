```python
import pandas as pd
```

```python
df = pd.read_csv("airline_delay_cause.csv")
```

```python
fact_flights = df[[
    'year','month','carrier','airport',
    'arr_flights','arr_delay',
    'carrier_delay','weather_delay',
    'nas_delay','security_delay','late_aircraft_delay'
]]

dim_carrier = df[['carrier','carrier_name']].drop_duplicates()
dim_airport = df[['airport','airport_name']].drop_duplicates()
```

```python
fact_flights.to_sql("fact_flights", engine, if_exists="replace", index=False)
dim_carrier.to_sql("dim_carrier", engine, if_exists="replace", index=False)
dim_airport.to_sql("dim_airport", engine, if_exists="replace", index=False)
```
